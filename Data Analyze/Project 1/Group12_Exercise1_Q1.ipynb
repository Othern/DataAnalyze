### 1.1. Perform an inner join on the "Car_Merge_A'' and "Car_Merge_B" datasets

- Using "car_ID" as the primary key.
- Then, concatenate the merged dataset with the "Car_Concat" dataset along the row axis.
- Convert any character columns to a factor in R and categorical in Pythons.

*Note that the remaining 'car_ID' is not one of our variables, delete it or convert it into the index.*


```python
# Inner join "Car_Merge_A" and "Car_Merge_B" datasets using "car_ID" as the primary key
merged_AB = pd.merge(car_merge_A, car_merge_B, on="car_ID")
merged_AB.head()
```

```python
# Concatenate the merged dataset with "Car_Concat" along the row axis
final_dataset = pd.concat([merged_AB, car_concat])
final_dataset
```

```python
# Convert any character columns to categorical in Python
final_dataset["drivewheel"] = final_dataset["drivewheel"].astype('category')
final_dataset.info()
```

```python
# To drop "car_ID":
final_dataset.drop("car_ID", axis=1, inplace=True)

# Display the first few rows of the final dataset to verify the changes
final_dataset.head()
```

### 1.3. Consider doing a series of bivariate analyses on "PRICE vs. the rest of variables."

Specifically, plot your data and perform bivariate statistical tests to understand the relationships among the variables

```python
continuous_vars = ['wheelbase', 'enginesize', 'boreratio', 'stroke', 'compressionratio', 'horsepower', 'peakrpm', 'mpg', 'PRICE']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

def plot_bivariate_analyses(current_index: int, min_x: float=-1, max_x: float=-1, min_y: float=-1, max_y: float=-1, use_log_x: bool=False, use_log_y: bool=False) -> None:
    """
    Create a scatter plot with a regression line and calculate Pearson correlation coefficients.

    Parameters:
    - current_index (int): Index of the continuous variable to be plotted against 'PRICE'.
    - min_x (float, optional): Minimum value for the x-axis (default is -1, auto-scale).
    - max_x (float, optional): Maximum value for the x-axis (default is -1, auto-scale).
    - min_y (float, optional): Minimum value for the y-axis (default is -1, auto-scale).
    - max_y (float, optional): Maximum value for the y-axis (default is -1, auto-scale).
    - use_log_x (bool, optional): Set to True to apply logarithmic scale to the x-axis (default is False).
    - use_log_y (bool, optional): Set to True to apply logarithmic scale to the y-axis (default is False).

    This function creates a scatter plot of the selected continuous variable against 'PRICE'. It also adds a regression line
    to visualize the linear relationship between the variables. The y-axis is inverted to have the origin at the bottom.
    Optional parameters allow for custom axis limits and logarithmic scaling.

    After plotting, the function calculates the Pearson correlation coefficient and p-value to quantify the linear
    relationship between the variables and displays them.

    Example usage:
    plot_bivariate_analyses(0) # Plot the first continuous variable against 'PRICE'.

    """
    # Create a scatter plot for the selected continuous variable vs. 'PRICE'
    plt.figure(figsize=(8, 6))
    ax = sns.scatterplot(data=final_dataset, x=continuous_vars[current_index], y='PRICE', color='blue')

    # Add a regression line to visualize the linear relationship
    sns.regplot(data=final_dataset, x=continuous_vars[current_index], y='PRICE', scatter=False, ax=ax, color='red')

    # Invert the y-axis to have the origin at the bottom
    ax.invert_yaxis()

    # Set the title and axis labels
    plt.title(f'Scatter Plot: {continuous_vars[current_index]} vs. PRICE')
    plt.xlabel(f'{continuous_vars[current_index]}')
    plt.ylabel('PRICE')

    # Set optional limits for the x and y axes
    if min_x != -1 and max_x != -1:
        plt.xlim(min_x, max_x)

    if min_y != -1 and max_y != -1:
        plt.ylim(min_y, max_y)

    # Apply logarithmic scale to the x and/or y-axis if specified
    if use_log_x:
        plt.xscale('log')

    if use_log_y:
        plt.yscale('log')

    # Display the plot
    plt.show()

    # Calculate the Pearson correlation coefficient and p-value
    correlation, p_value = pearsonr(final_dataset[continuous_vars[current_index]], final_dataset['PRICE'])

    # Display the correlation coefficient and p-value
    print(f"Pearson Correlation Coefficient: {correlation:.2f}")
    print(f"P-Value: {p_value:.8f}")
```

### 1.4. Please perform normality tests on PRICE.
- Does it seem "normal?"
- If not, do you think fitting general linear models to predict or explain the outcome is appropriate?


```python
import matplotlib.pyplot as plt
from scipy.stats import shapiro, probplot

# Extract the PRICE column
price_data = final_dataset['PRICE']

# Perform Shapiro-Wilk normality test
stat, p = shapiro(price_data)

# Interpret the results
alpha = 0.05
if p > alpha:
    print("PRICE is normally distributed (p =", p, ")")
else:
    print("PRICE is not normally distributed (p =", p, ")")

# Visualize the data using a histogram and Q-Q plot
plt.figure(figsize=(12, 5))

# Histogram
plt.subplot(1, 2, 1)  # Creating a subplot, one row, two columns, first plot
plt.hist(price_data, bins=20, color='blue', alpha=0.7)  # Creating a histogram
plt.title('Histogram of PRICE')  # Adding a title to the histogram
plt.xlabel('Price')  # Labeling the x-axis
plt.ylabel('Frequency')  # Labeling the y-axis

# Q-Q plot
plt.subplot(1, 2, 2)  # Creating the second plot
probplot(price_data, plot=plt)  # Creating a Q-Q plot
plt.title('Q-Q Plot of PRICE')  # Adding a title to the Q-Q plot

plt.tight_layout()  # Ensuring proper spacing between subplots
plt.show()  # Displaying the plots
```

### 1.5. Consider fitting linear models with manually selected variables (i.e., multivariate analysis).
- What is your best model?
- You may consider those variables with "p < 0.05".

```python
from sklearn.model_selection import train_test_split
import pandas as pd

# Set a random seed
random_seed = 20230929

# Specify the features (X) and the target variable (y)
X = final_dataset.drop(columns=['PRICE'])  # Features
y = final_dataset['PRICE']  # Target variable

# Perform one-hot encoding for categorical variables
X_encoded = pd.get_dummies(X, columns=['drivewheel'], drop_first=True)

# Split the dataset into a 70% training set and a 30% testing set
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.30, random_state=random_seed)

# Print the shapes of the training and testing sets to verify the split
print("Training set shape (X, y):", X_train.shape, y_train.shape)
print("Testing set shape (X, y):", X_test.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Create a LinearRegression model
model = LinearRegression()

# Manually select the predictor variables
selected_features = ['wheelbase', 'enginesize', 'boreratio', 'compressionratio', 'horsepower', 'mpg']

# Extract the selected features from the training and testing sets
X_train_selected = X_train[selected_features]
X_test_selected = X_test[selected_features]

# Fit the model to the training data using the selected features
model.fit(X_train_selected, y_train)

# Make predictions on the training data
y_train_pred = model.predict(X_train_selected)

# Make predictions on the testing data
y_test_pred = model.predict(X_test_selected)

# Evaluate the model's performance on the training data
train_mse = mean_squared_error(y_train, y_train_pred)
train_r2 = r2_score(y_train, y_train_pred)

# Evaluate the model's performance on the testing data
test_mse = mean_squared_error(y_test, y_test_pred)
test_r2 = r2_score(y_test, y_test_pred)

# Store the result
model_result = {"manually selected linear model" : {"train": y_train_pred, "test": y_test_pred}}

# Print the model's performance metrics
print("Training MSE:", train_mse)
print("Training R-squared:", train_r2)
print("\nTesting MSE:", test_mse)
print("Testing R-squared:", test_r2)
```

### 1.7. Computes Root Mean Square Error (RMSE), which is defined as:
![Root Mean Square Error](RootMeanSquareError.png)

- where and are actual and predicted values, respectively.
- Then apply your linear models to the training and testing sets, train_d and test_d.
- What are the RMSEs of your models?
- What is your best model in terms of accuracy of prediction (with lowest RMSE)?

```python
#Define a function to calculate RMSE (Root Mean Square Error)
import numpy as np
from typing import Union

def calculate_root_mean_square_error(actual: Union[list, np.ndarray], predicted: Union[list, np.ndarray]) -> float:
    """
    Calculate the Root Mean Square Error (RMSE) between actual and predicted values.

    Parameters:
    - actual (list or np.ndarray): The actual target values.
    - predicted (list or np.ndarray): The predicted target values.

    Returns:
    - rmse (float): The Root Mean Square Error.
    """
    # Convert input lists/arrays to numpy arrays to ensure compatibility
    actual = np.array(actual)
    predicted = np.array(predicted)

    # Calculate the squared differences between actual and predicted values
    squared_errors = (actual - predicted) ** 2

    # Calculate the mean of squared errors
    mse = squared_errors.mean()

    # Calculate the square root of the mean squared error to get RMSE
    return np.sqrt(mse)
```

```python
# Evaluate the model's performance on the training data
train_rmse_1 = calculate_root_mean_square_error(y_train, model_result["manually selected linear model"]["train"])

# Evaluate the model's performance on the testing data
test_rmse_1 = calculate_root_mean_square_error(y_test, model_result["manually selected linear model"]["test"])

# Print the model's performance metrics
print("Training RMSE:", train_rmse_1)
print("\nTesting RMSE:", test_rmse_1)
```

```python
# Evaluate the model's performance on the training data
train_rmse_2 = calculate_root_mean_square_error(y_train, model_result["all selected linear model"]["train"])

# Evaluate the model's performance on the testing data
test_rmse_2 = calculate_root_mean_square_error(y_test, model_result["all selected linear model"]["test"])

# Print the model's performance metrics
print("Training RMSE:", train_rmse_2)
print("\nTesting RMSE:", test_rmse_2)
```
